```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox


root = Tk()
root.title('Library Book Management System')

```

```python
def Database():
    global conn,cursor
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS books(book_id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,book_title TEXT,author TEXT,year INTEGER,isbn INTEGER)")
    
    
def Create():
    if BOOKTITLE.get() =="" or AUTHOR.get()=="" or YEAR.get()=="" or ISBN.get()=="":
        txt_result.config(text="Please enter all",fg ="red")
    else:
        Database()
        cursor.execute("INSERT INTO books(book_title,author,year,isbn) VALUES(?,?,?,?)",(str(BOOKTITLE.get()),str(AUTHOR.get()),int(YEAR.get()),int(ISBN.get())))
        conn.commit()
        BOOKTITLE.set("")
        AUTHOR.set("")
        #YEAR.set("")
        #ISBN.set("")
        cursor.close()
        conn.close()
        txt_result.config(text="Record Inserted !",fg = "green")
        
              
def Read():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM books ORDER BY book_title  ASC")
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
        cursor.close()
        txt_result.config(text = "Successfully fetched data !",fg="black")
           
def Search():
    tree.delete(*tree.get_children())
    Database()
    #cursor.execute("SELECT * FROM books WHERE book_title =? OR author = ? OR year=? OR isbn=?",(str(BOOKTITLE.get()),str(AUTHOR.get()),str(YEAR.get()),str(ISBN.get())))
    if str(BOOKTITLE.get())!="" :
        #name=str(BOOKTITLE.get())
        #temp='SELECT * FROM books WHERE BOOKTITLE='+name
        cursor.execute("SELECT * FROM books WHERE book_title=?",(str(BOOKTITLE.get()),))
        fetch = cursor.fetchall()
        for data in fetch:
            tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
            #cursor.close()
            #conn.close()
            #BOOKTITLE.set("")
            #AUTHOR.set("")
            #YEAR.set("")
            #ISBN.set("")
        txt_result.config(text = "Successfully Searched data !",fg="green")
    elif str(AUTHOR.get())!="":
        #name=" ' "+str(AUTHOR.get()).title+" ' "
        #emp='SELECT * FROM books WHERE AUTHOR='+name
        cursor.execute("SELECT * FROM books WHERE author=?",(str(AUTHOR.get()),)) 
        fetch = cursor.fetchall()
        for data in fetch:
             tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
                
        txt_result.config(text = "Please see the books !",fg="green")
        
    elif str(YEAR.get())!="":
        #emp='SELECT * FROM books WHERE YEAR='+str(YEAR.get())
        cursor.execute("SELECT * FROM books WHERE year=?",(str(YEAR.get()),)) 
        fetch = cursor.fetchall()
        for data in fetch:
             tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
                
        txt_result.config(text = "Please see the books !",fg="green")
        
    elif str(ISBN.get())!="":
        #temp='SELECT * FROM books WHERE ISBN='+str(ISBN.get())
        cursor.execute("SELECT * FROM books WHERE isbn=?",(str(ISBN.get()),)) 
        fetch = cursor.fetchall()
        for data in fetch:
             tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
                
        txt_result.config(text = "Please see the books !",fg="green") 
    else:
        txt_result.config(text = "Please enter atleast single identity !",fg="red") 
    conn.close()
    
    
    
def Update():
    Database()
    cursor.execute("UPDATE books SET author=?,year=?,isbn=? WHERE book_title=?",(str(AUTHOR.get()),int(YEAR.get()),int(ISBN.get()),str(BOOKTITLE.get())))
    conn.commit()
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[1],data[2],data[3],data[4]))
    BOOKTITLE.set("")
    AUTHOR.set("")
    txt_result.config(text = "Successfully updated!",fg="green")
    conn.close()
    
    
def Issue():
    Database()
    selected_item=tree.selection()
    if len(selected_item)>0:
        tree.delete(selected_item)
        txt_result.config(text = "Selected book is issued !",fg="green") 
    else:
        txt_result.config(text = "Please select book from library to issue !",fg="red") 
    conn.close()
         
    
def Delete():
    Database()
    cursor.execute("DELETE FROM books WHERE book_title=? AND author=?AND year=? AND isbn=?",(str(BOOKTITLE.get()),str(AUTHOR.get()),int(YEAR.get()),int(ISBN.get())))
    conn.commit()
    conn.close()
    BOOKTITLE.set("")
    AUTHOR.set("")
    txt_result.config(text = "Successfully deleted!",fg="red")
            
def Exit():
    result = tkMessageBox.askquestion('Do you want to exit ? (y/n)',icon='warning')
    if result == 'yes':
        root.destroy()
        exit()
    
```

```python
#variable Declaration

BOOKTITLE = StringVar()
AUTHOR = StringVar()
YEAR = StringVar()
ISBN = StringVar()
```

```python
# Frame

Top = Frame(root, width=900, height=50, bd=8,relief='raise',bg='blue')
Top.pack(side=TOP)
Left = Frame(root, width=600, height=500, bd=8,relief='raise',bg='red')
Left.pack(side=LEFT)
Right = Frame(root, width=600, height=400, bd=8,relief='raise',bg='yellow')
Right.pack(side=RIGHT)


Forms= Frame(Left,width=600,height=450)
Forms.pack(side=TOP)

Buttons=Frame(Left,width=600,height=100,bd=8,relief='raise')
Buttons.pack(side=BOTTOM)





```

```python
# Label widget

txt_title=Label(Top,width=900,font = ('arial',24),text='Library Management Application')
txt_title.pack()
txt_Booktitle=Label(Forms,text="Booktitle:",font = ('arial',16),bd=15)
txt_Booktitle.grid(row=0,stick="e")
txt_Author=Label(Forms,text="Author:",font = ('arial',16),bd=15)
txt_Author.grid(row=1,stick="e")
txt_Year=Label(Forms,text="Year:",font = ('arial',16),bd=15)
txt_Year.grid(row=2,stick="e")
txt_ISBN=Label(Forms,text="ISBN:",font = ('arial',16),bd=15)
txt_ISBN.grid(row=3,stick="e")
txt_result = Label(Buttons)
txt_result.pack(side=TOP)


# Entry widget




Booktitle = Entry(Forms,textvariable=BOOKTITLE,width=30)
Booktitle.grid(row=0,column=1)
Author = Entry(Forms,textvariable=AUTHOR,width=30)
Author.grid(row=1,column=1)
Year = Entry(Forms,textvariable=YEAR,width=30)
Year.grid(row=2,column=1)
ISBN = Entry(Forms,textvariable=ISBN,width=30)
ISBN.grid(row=3,column=1)


# Button widget


btn_Display = Button(Buttons, width=10, text="Display",command=Read)
btn_Display.pack(side=LEFT)
btn_search = Button(Buttons, width=10, text="Search",command=Search)
btn_search.pack(side=LEFT)
btn_add = Button(Buttons, width=10, text="Add",command=Create)
btn_add.pack(side=LEFT)
btn_update = Button(Buttons,width=10,text="Update",command=Update)
btn_update.pack(side=LEFT)
btn_issue = Button(Buttons, width=10, text="Issue",command=Issue)
btn_issue.pack(side=LEFT)
btn_delete = Button(Buttons, width=10, text="Delete",command=Delete)
btn_delete.pack(side=LEFT)
btn_exit = Button(Buttons, width=10, text="Exit",command=Exit)
btn_exit.pack(side=LEFT)




# List widget

scrollbary = Scrollbar(Right,orient=VERTICAL)
scrollbarx = Scrollbar(Right,orient=HORIZONTAL)


tree = ttk.Treeview(Right,columns=("Booktitle","Author","Year","ISBN"))


scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM,fill=X)
tree.heading('Booktitle',text="Booktitle",anchor=W)
tree.heading('Author',text="Author",anchor=W)
tree.heading('Year',text="Year",anchor=W)
tree.heading('ISBN',text="ISBN",anchor=W)
tree.column('#0',stretch=NO,minwidth=0,width=0)
tree.column('#1',stretch=NO,minwidth=0,width=80)
tree.column('#2',stretch=NO,minwidth=0,width=120)
tree.column('#3',stretch=NO,minwidth=0,width=80)
tree.column('#4',stretch=NO,minwidth=0,width=150)
tree.pack()


# Initialization


if __name__ == '__main__':
    root.mainloop()



```
